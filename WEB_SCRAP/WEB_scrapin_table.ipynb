```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_banks"
# obtener el contenido de la página web en formato de texto 
data  = requests.get(url).text
```

```python
data[101:124]
```

### Scrap de los datos wikipedia

 Utilizando  `BeautifulSoup`, se cargan los datos de la tabla `By_market_capitalization` en un  `txt`.

```python
soup = BeautifulSoup(data,"html.parser")

foundtext = soup.find(id="By_market_capitalization")
txtable = foundtext.findNext('table')
txtable
```

```python
# encontrar todas las tablas html en la página web
tables = soup.find_all('table') # en la tabla html está representado por la etiqueta <table>
# numero de tablas por la longitud de la lista
len(tables)
```

se carga la tabla `By market capitalization`  a un pandas dataframe. 


```python
pd.read_html(str(tables[2]), flavor='bs4')
Table = pd.read_html(str(tables[3]), flavor='bs4')[0]

Table
```
