```python
from bs4 import BeautifulSoup 
import requests  
```

```python
url = "http://www.ibm.com"
```

```python
data  = requests.get(url).text 
```

```python
soup = BeautifulSoup(data,"html.parser")  # crea un objeto de sopa usando la variable 'data'
```

## Scrap de todos los links


```python
for link in soup.find_all('a',href=True):  # en html los links está representado por la etiqueta <a>

    print(link.get('href'))
```

## Scrap de todas las imágenes Etiquetas


```python
for link in soup.find_all('img'):# en html la imagen está representada por la etiqueta <img>
    print(link)
    print(link.get('src'))
```

## Tablas de wikipedia a pandas


```python
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/World_population"
```

```python
# obtener el contenido de la página web en formato de texto 
data  = requests.get(url).text
```

```python
soup = BeautifulSoup(data,"html.parser")
```

```python
# encontrar todas las tablas html en la página web
tables = soup.find_all('table') # en la tabla html está representado por la etiqueta <table>
```

```python
# numero de tablas por la longitud de la lista
len(tables)
```

```python
for index,table in enumerate(tables):
    if ("10 most densely populated countries" in str(table)):# se toma una tabla
        table_index = index
print(table_index)
```

```python
print(tables[table_index].prettify())
```

```python
population_data = pd.DataFrame(columns=["Clasificación", "Pais", "Poblacion", "Area", "Densidad"]) #estructura del la tabla

for row in tables[table_index].tbody.find_all("tr"):#iteraciones para rellenar la tabla 
    col = row.find_all("td")
    if (col != []):
        rank = col[0].text
        country = col[1].text
        population = col[2].text.strip()
        area = col[3].text.strip()
        density = col[4].text.strip()
        population_data = population_data.append({"Clasificación":rank, "Pais":country, "Poblacion":population, "Area":area, "Densidad":density}, ignore_index=True)

population_data
```
